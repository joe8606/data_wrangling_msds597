```python
def scrape_reviews(df):
    # Create an empty list to store the reviews data
    reviews_data = []

    # Iterate through each URL in the dataframe
    for index, row in df.iterrows():
        url = row['url']
        try:
            # Make a GET request to fetch the page content
            response = requests.request("GET", url, headers=headers, data=payload)
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract reviews
            review_divs = soup.find_all('div', {'data-hb-id': 'BoxV3', 'data-rtl-id': lambda x: x and x.startswith('reviewCard-')})
            reviews = []
            for div in review_divs:
                review_text = div.find('p', class_='_6o3atzbl _6o3atz174')
                if review_text:
                    reviews.append(review_text.text)

            # Append the URL and reviews to the reviews_data list
            reviews_data.append([url] + reviews)
            

        except Exception as e:
            print(f"Error scraping URL {url}: {e}")
            reviews_data.append([url])  # Append only the URL if scraping fails
    print(reviews_data)
    # Create a new dataframe for reviews
    max_reviews = max(len(row) for row in reviews_data)  # Find the maximum number of reviews
    columns = ['url'] + [f'review_{i+1}' for i in range(max_reviews - 1)]
    reviews_df = pd.DataFrame(reviews_data, columns=columns)

    return reviews_df

# Example usage
reviews_df = scrape_reviews(df)
print(reviews_df)
```
